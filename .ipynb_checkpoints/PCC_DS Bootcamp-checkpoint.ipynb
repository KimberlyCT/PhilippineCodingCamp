### Sum of Even numbers
>Using:
> 1. While
> 2. For


```python
numIn = int(input("Give me a number:"))

if numIn <= 2: print(numIn)

else: 
    xStart = 2
    numInSum = 0
    while(xStart <= numIn and xStart%2 == 0):
        numInSum = numInSum + xStart
        xStart = xStart + 2
    print("Sum of even numbers under given number: ",numInSum) 
```

```python
upperLimit = int(input("Enter a number: "))
sumOfEvents = 0

for number in range (2, upperLimit + 1, 2):
        sumOfEvents += number
print("The sum of even numbers is: ", sumOfEvents)
```


### Create Simple Calculator

```python
###Simple Calculator
import operator

while input("Simple Calculator, press enter to start, type ""q"" to exit.") != "q":
    try:
        
        operand1 = int(input("Input 1st number: "))
        operand2 = int(input("Input 2nd number: "))
        operation = {"+": operator.add, "-": operator.sub, "/": operator.truediv, "*": operator.mul}
        operationSymbol = str(input("Choose operation (+,-,/,*)"))
        
        for i in operation:
            if operationSymbol == i:
                print("Answer: ",operation[i](operand1,operand2))
                break
            else:
                continue
    except IndexError:
        print("Invalid Operator")
    except ValueError: 
        print("Error")

        
print("Calculator has stopped.")
```

## Practice sets Week 8

##### Problem Statement: To Predict if the customer will repay loan amount or not. Use Decision tree algorithm in python.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

df = pd.read_csv('Decision_Tree_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
X = df.values [:, 0:4]
X
```

```python
y = df.values[:,5]
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state = 100)
clf = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth = 3, min_samples_leaf = 5 )
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
score
```

```python
print('Accuracy score is', score)
```

```python
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names = X, class_names = clf.classes_)
plt.show
```

### Practice Application

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
X = df.values [:, 0:7]
y = df.values[:,8]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state = 100)
clf = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth = 3, min_samples_leaf = 5 )
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
print('Accuracy score is', score)
```

```python
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names = X, class_names = ['No', 'Yes'])
plt.show


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


df= pd.read_csv('homeprices.csv')
df
reg = LinearRegression()
reg.fit(df[['area']],df['price'] )
d = pd.read_csv('areas.csv')
predictions = reg.predict(d[['area']])
d['predicted_price'] = predictions
d
```

```python
plt.scatter(df['area'], df['price'], color ='blue', marker = 'o', label = 'Actual Prices')
plt.plot(df['area'], reg.predict(df[['area']]), color ='red', label='Regression Line')
plt.scatter(d['area'], d['predicted_price'], color ='green', marker = 'x', label = 'Predicted Prices')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Home Prices Prediction')
plt.legend()
plt.show()
```
