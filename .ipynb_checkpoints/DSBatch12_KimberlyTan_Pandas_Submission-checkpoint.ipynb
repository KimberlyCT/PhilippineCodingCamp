## 5. Count total cars per company
### Checking & Solution:

```python
TotalCars = cars[["company"]]
Sumcars = TotalCars.groupby(by='company')['company'].count()
Sumcars
##SumCars = TotalCars.groupby(by = 'company', as_index = False).count()
##SumCars
```

## 6.Find each company’s Highest priced car
### Checking & Solution:

```python
CompPrice = cars[["company","price"]]
Pricesort = CompPrice.sort_values(by=["company","price"],ascending = False)
Pricesort 
```

```python
HiPrice = Pricesort.groupby(by="company").first()
HiPrice
```

## 7. Find the average mileage of each car making company
### Checking & Solution:

```python
cars.info()
```

```python
Milecar = cars[["company","average-mileage"]]
AveMile = Milecar.groupby(["company"]).mean("average-mileage")
AveMile
```
