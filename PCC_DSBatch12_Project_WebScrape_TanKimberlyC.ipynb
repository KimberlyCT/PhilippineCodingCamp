```python
from IPython.display import display, HTML
from bs4 import BeautifulSoup as bs
from urllib.parse import urljoin
import requests as rq
import pygal

url = 'https://books.toscrape.com/index.html'

page = rq.get(url).text

soup = bs(page,'html.parser')
```

<h2>Solution</h2>

<h3>Data Collection</h3>

```python
categoryHolder =soup.find('ul',class_="nav nav-list")
categoryItems= categoryHolder.find_all('a')
```

```python
categoryItems
```

<h4>Book Genres</h4>

```python
categories = [link.text.strip() for link in categoryItems]
Genres = categories[1:]
print(Genres)
```

<h4>Books per Genre</h4>

```python
GenreLinks = [link['href'] for link in categoryItems]
GenreLinks
```

```python
baseURL = 'https://books.toscrape.com/'
CompleteGenreLinks = [urljoin(baseURL,link) for link in GenreLinks[1:]]
CompleteGenreLinks
```

```python
BooksPerGenre=[]
for link in CompleteGenreLinks:
    subURL = link
    subPage = rq.get(subURL).text
    subSoup = bs(subPage,'html.parser')
    getForm= subSoup.find(attrs={"class":"form-horizontal","method":"get"})
    getResult = getForm.find('strong').text.strip()
    BooksPerGenre.append(getResult)

BooksPerGenre
```

```python
import numpy as np
import math
PagesPerGenre = []
for inventory in BooksPerGenre:
    PagesPerGenre.append(int(np.ceil(int(inventory)/20)))
PagesPerGenre
```

```python
print(len(Genres),len(BooksPerGenre),len(PagesPerGenre))
```

<h4>List of Book Details Per Genre</h4>
<h5> Includes Book Titles and their corresponding Genre, Price, and Ratings</h5>

```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
from urllib.parse import urljoin

# Create an empty DataFrame for the genre
booksOfGenres = pd.DataFrame(columns=["Book Titles", "Genre"])

for linkCounter, genre in enumerate(Genres):
    
    # Fetch the first page for the current genre
    genURL = CompleteGenreLinks[linkCounter]
    genPage = rq.get(genURL).text
    
    # Find all book titles on the first page & populate dedicated DataFrame
    genSoup = bs(genPage, 'html.parser')
    BookHolders = genSoup.find_all('h3')
    bookNames = [link.a['title'] for link in BookHolders]
    
    # Do the same for Ratings & Prices
    rateHolders = genSoup.find_all(attrs={'class':'star-rating'})  
    bookRatings = [rate['class'][1] for rate in rateHolders] 

    priceHolders = genSoup.find_all(attrs='price_color')
    bookPrices = [price.text for price in priceHolders]

    
    # Create a DataFrame from the book titles and genre
    genreData = pd.DataFrame({
        "Book Titles": bookNames,
        "Genre": [genre] * len(bookNames),  # Repeating the genre for each book
        "Rating": bookRatings,
        "Price": bookPrices
        
    })
    
    # Concatenate with the main DataFrame
    booksOfGenres = pd.concat([booksOfGenres, genreData], ignore_index=True)

    # Check if there are more pages
    pageCounter = 1
    while pageCounter != PagesPerGenre[linkCounter]:
        genNext = genSoup.find(attrs={"class": "next"})
        if not genNext:
            break  # If there is no 'next' button, exit the loop
        
        # Get the link to the next page
        basegenURL = CompleteGenreLinks[linkCounter].rpartition('/')[0] + '/'
        nextLink = urljoin(basegenURL, genNext.a['href'])
        
        # Fetch the next page
        nextPage = rq.get(nextLink).text
        nextSoup = bs(nextPage, 'html.parser')
        nextBooks = nextSoup.find_all('h3')
        nextBookNames = [link.a['title'] for link in nextBooks]

        nextPriceHolders = nextSoup.find_all(attrs='price_color')
        nextbookPrices = [price.text for price in nextPriceHolders]

        nextrateHolders = nextSoup.find_all(attrs={'class':'star-rating'})
        nextbookRatings = [rate['class'][1] for rate in nextrateHolders] 
        
        # Create DataFrame for the next page and append
        nextPageData = pd.DataFrame({
            "Book Titles": nextBookNames,
            "Genre": [genre] * len(nextBookNames),
            "Rating": nextbookRatings,
            "Price": nextbookPrices
            
        })
        
        booksOfGenres = pd.concat([booksOfGenres, nextPageData], ignore_index=True)
        
        # Update soup and page counter
        genSoup = nextSoup
        pageCounter += 1
booksOfGenres
```

<h3>Data Visualization</h3>

```python
toRep= {'Rating':{'One','Two','Three','Four','Five'}}
valuesRep = {'Rating':{1,2,3,4,5}}

rateSummary= booksOfGenres.groupby(['Rating'],as_index=False).agg({'Book Titles':'count'}).replace(to_replace = toRep, value = valuesRep).sort_values(by='Rating',ascending=False)
rateSummary
```

<h4>Books per Rating: Bar Chart</h4>

```python
import matplotlib.pyplot as plt
colors = ['#7BD3EA','#A1EEBD','#FFD09B','#FFE6E6','#FFB0B0',]
ratesBar = rateSummary.plot(x='Rating', y='Book Titles', kind='bar', color = colors, legend=False)


# Add labels and title
plt.xlabel('Rating')
plt.ylabel('Number of Books')
plt.title('Number of Books per Rating')
plt.xticks(rotation=0)

# Annotate each bar with its height (book count)
for p in ratesBar.patches:
    ratesBar.annotate(str(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Position
                ha='center', va='center', 
                xytext=(0, 5), textcoords='offset points')  # Offset to avoid overlap
plt.gca().set_facecolor('azure')



# Show the plot
plt.show()
```

```python
genreSummary= booksOfGenres.groupby(['Genre'],as_index=False).agg({'Book Titles':'count'}).sort_values(by='Book Titles',ascending=False)
genreSummary['Portion size'] = (genreSummary[['Book Titles']])/(genreSummary[['Book Titles']].sum())*100
genreSummary['Portion category'] = np.where(genreSummary['Portion size'] < 2, '< 2%', round(genreSummary['Portion size'],2))
genreSummary['Genres categorized'] = np.where(genreSummary['Portion size'] < 2, '+38 Others (under 20 books)', genreSummary['Genre'])
genreSummary
```

```python
print(genreSummary.groupby(['Genres categorized'],as_index=False).agg({'Book Titles':'count'}))
```

```python
GenrePieSummary= genreSummary.groupby(['Genres categorized'],as_index=False).agg({'Book Titles':'sum'}).sort_values(by='Book Titles',ascending=False)
GenrePieSummary
```

<h4>Genres: Pie Chart</h4>

```python
import matplotlib.pyplot as plt
import numpy as np

colors = ['#CFCFCF', '#8E7AB5' ,'#AD88C6','#E1AFD1','#FFE6E6','#FFB0B0','#FFD09B','#A1D6B2','#CEDF9F','#E8B86D','#E8C5E5','#C7B7A3','#4F6F52']
#explosion
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)

# Create a larger figure
fig2, genreOverview = plt.subplots(figsize=(20, 18), subplot_kw=dict(aspect="equal"))

# Define a function to display both genre and percentage/absolute count
def func(pct, allvals, genre):
    absolute = int(np.round(pct / 100. * np.sum(allvals)))
    return f"{genre}{pct:.1f}%\n({absolute:d} books)"

# Assuming genreSummary is a DataFrame with 'Book Titles' and 'Genre' columns
wedges, texts, autotexts = genreOverview.pie(
    GenrePieSummary['Book Titles'], 
    autopct=lambda pct: func(pct, GenrePieSummary['Book Titles'], ""),  # Temporarily use an empty string for genre
    textprops=dict(color='#3B3030'),
    pctdistance=0.65, 
    labeldistance=1.1,
    colors = colors,
    explode = explode,
    startangle = 90,
    
)


# Manually set the labels inside the wedges
for i, wedge in enumerate(wedges):
    genre = GenrePieSummary['Genres categorized'][i]
    # Use the genre name in the corresponding wedge's text
    autotexts[i].set_text(f"{genre}\n\n\n{autotexts[i].get_text()}")


# genreOverview.legend(wedges, GenrePieSummary['Genres categorized'],
#                      title="Genres",
#                      loc="center left",
#                      borderpad=0.5,
#                      bbox_to_anchor=(0.9, 0.2, 0.5, 1),
#                      fontsize = 14,
#                      )
# Customize text size and weight
plt.setp(autotexts, size=12, weight="bold")

# Add a title
genreOverview.set_title("Books Collection", fontsize= 30, y=0.93, fontweight='bold')
# Show the chart

genreOverview
```
